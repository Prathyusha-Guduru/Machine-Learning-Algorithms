### Question 2 
Euclidean distance, (ii) Manhattan distance and (iii) Minkowski distance Matrices* for all the data samples in the feature space. 


```python
import pandas as pd 
from sklearn import datasets 
iris_dataset = datasets.load_iris() 
iris_dataset.keys() 
```

```python
iris = pd.DataFrame(iris_dataset.data, 
columns=iris_dataset.feature_names) 
iris["species"] = iris_dataset.target 
iris.head() 
```

```python
np_iris= iris.to_numpy() 
np_iris
```

```python
def euclidean_distance(p,q): 
	dist = 0 
	dist = dist + (p - q)**2 
	euclidean_dist = np.sqrt(dist) 
	return euclidean_dist 
distance_list = [] 
for x in np_iris: 
	for y in x: 
		distance = euclidean_distance(x,y) 
		distance_list.append(distance)
print("Euclidean distance: ") 
print(distance_list) 

```

```python
def minikowski_distance(p, q): 
	dist = 0 
	dist = dist + np.power(np.abs(p - q),3) 
	minikowski_dist = dist 
	return minikowski_dist 
distance_list2 = [] 
for x in np_iris: 
	for y in x: 
		distance = minikowski_distance(x,y) 
		distance_list2.append(distance)
print("Minikowski distance of IRIS DATASET: ")
print(distance_list2)
```
